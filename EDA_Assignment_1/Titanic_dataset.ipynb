### Fill Missing Ages
Replace missing values in the `Age` column with the median age.

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

###  Drop Irrelevant Columns
Remove `PassengerId`, `Ticket`, and `Cabin` as they provide little or redundant information.

```python
df.drop(['PassengerId', 'Ticket', 'Cabin'], axis=1, inplace=True)
```

###  Fill Missing Embarked Values
Replace missing values in `Embarked` with the mode (most common port).
Replace missing `Fare` values with the mean fare.

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)
```

###  Encode Survived as Categorical
Convert the binary values of `Survived` (0/1) to string labels ("Died", "Survived").
Also convert to `category` dtype.

```python
df["Survived"]=df["Survived"].map({0:"Died",1:"Survived"})
df['Survived'] = df['Survived'].astype('category')
```

###  Convert Types
- Convert `Pclass` to categorical (ordinal).
- Convert `Age` to integer for simplicity.

```python
df['Pclass'] = df['Pclass'].astype('category')
df['Age'] = df['Age'].astype('int64')
```

```python
df.info()
```

###  Median Age by Sex and Class
Group passengers by `Sex` and `Pclass` to explore patterns in median age.

```python
Age_mean=df.groupby(['Sex', 'Pclass']).median(numeric_only=True)['Age']
Age_mean
```

###  Extract Titles from Names
Create a new feature `Title` by extracting honorifics (e.g., Mr, Miss, Mrs) from the `Name` column.

```python
df['Title']=df['Name'].str.split(", ",expand=True)[1].str.split(".",expand=True)[0]
```

### Simplify and Group Rare Titles
Group rare or less frequent titles into a 'Rare' category and fix similar titles.

```python

df['Title'] = df['Title'].replace(['Lady', 'the Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
df['Title'] = df['Title'].replace('Mlle', 'Miss')
df['Title'] = df['Title'].replace('Ms', 'Miss')
df['Title'] = df['Title'].replace('Mme', 'Mrs')
df['Title'].value_counts()
```

###  Drop Name Column
The useful information (title) has been extracted, so we drop the original `Name` column.


```python
df.drop(['Name'], axis=1, inplace=True)
```

### Create Family Size Feature
Combine `SibSp` (siblings/spouses) and `Parch` (parents/children) to create a new `FamilySize` feature.

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df['Survived'].value_counts().plot(kind='bar', color=['tomato', 'skyblue'])
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

More passengers died than survived.

The dataset is imbalanced, with approximately 62% died and 38% survived.



```python
pclass_surv = pd.crosstab(df['Pclass'], df['Survived'])
pclass_surv.plot(kind='bar', stacked=True)
plt.title('Survival by Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.xticks(rotation=0)
plt.show()
```

Insight:

1st class passengers had a much higher survival rate than 2nd and 3rd classes.

3rd class passengers had the lowest chance of survival.

```python
sex_surv = pd.crosstab(df['Sex'], df['Survived'])
sex_surv.plot(kind='bar', stacked=True, color=['salmon', 'lightblue'])
plt.title('Survival by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Survived')
plt.show()
```

insight:

Females had a much higher survival rate than males.

The majority of those who died were males.

```python
embark_surv = pd.crosstab(df['Embarked'], df['Survived'])
embark_surv.plot(kind='bar', stacked=True)
plt.title('Survival by Embarked Port')
plt.xlabel('Port')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.xticks(rotation=0)
plt.show()
```

Insight:

Passengers from Cherbourg (C) had the highest survival proportion.

Embarked location appears to have some influence on survival.

```python
plt.hist(df[df['Survived'] == 'Survived']['Age'], bins=20, alpha=0.8, label='Survived', color='blue')
plt.hist(df[df['Survived'] == 'Died']['Age'], bins=20, alpha=0.5, label='Died', color='red')
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.show()
```

Insight:

Most deaths occurred among adults (20–40 years).

Children (especially below 10) had relatively higher survival.

```python
df.boxplot(column='Fare', by='Survived', grid=False)
plt.title('Fare Distribution by Survival')
plt.suptitle('')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()
```

Survivors generally paid higher fares.


```python
title_surv = pd.crosstab(df['Title'], df['Survived'])
title_surv.plot(kind='bar', stacked=True, figsize=(10, 5))
plt.title('Survival by Title')
plt.xlabel('Title')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.xticks(rotation=45)
plt.show()
```

Insight:

Titles like Mrs, Miss had higher survival rates.

Passengers with rare titles (officers, aristocrats) had better survival chances, suggesting social status mattered.

```python
labels = df['Survived'].value_counts().index
sizes = df['Survived'].value_counts().values
colors = ['lightcoral', 'lightgreen']

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Survival Distribution')
plt.axis('equal')
plt.show()
```

```python
avg_fare = df.groupby('Pclass')['Fare'].mean()
plt.plot(avg_fare.index, avg_fare.values, marker='o', linestyle='-', color='orange')
plt.title('Average Fare by Class')
plt.xlabel('Pclass')
plt.ylabel('Average Fare')
plt.xticks([1, 2, 3])
plt.grid(True)
plt.show()
```

There’s a strong inverse relationship between class number and fare.

1st class paid the most; 3rd class paid the least.



```python
df.boxplot(column='Age', by='Survived', grid=False)
plt.title('Age Distribution by Survival')
plt.suptitle('')
plt.xlabel('Survival Status')
plt.ylabel('Age')
plt.show()
```

Survivors had a slightly lower median age than those who died.

Survivors include more younger passengers and children.

```python
colors = df['Survived'].map({'Died': 'red', 'Survived': 'green'})

plt.figure(figsize=(8,6))
plt.scatter(df['Age'], df['Fare'], c=colors, alpha=0.6)
plt.title('Fare vs Age (colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

Insight:

High-fare older passengers (likely 1st class) had better survival.

Most deaths were clustered among young to middle-aged low-fare passengers.


```python
embarked_counts = df['Embarked'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(embarked_counts, labels=embarked_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
plt.title('Port of Embarkation Distribution')
plt.axis('equal')
plt.show()
```

```python
import numpy as np

numeric_df = df.select_dtypes(include=['int64', 'float64'])
correlation = numeric_df.corr()

fig, ax = plt.subplots(figsize=(8, 6))
cax = ax.matshow(correlation, cmap='coolwarm')

for (i, j), val in np.ndenumerate(correlation.values):
    ax.text(j, i, f'{val:.2f}', ha='center', va='center', color='black', fontsize=8)

fig.colorbar(cax)
plt.title('Correlation Heatmap with Values', pad=40)
plt.tight_layout()
plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=90)
plt.yticks(range(len(correlation.columns)), correlation.columns)
plt.xticks(rotation=0)
plt.show()
```

FamilySize is highly correlated with both SibSp and Parch, which confirms it's a derived feature.

Fare has low but positive correlation with all other features — passengers with larger families may have paid more, but not always.

Age has a negative correlation with family-related features — younger passengers were more likely to travel with family.
