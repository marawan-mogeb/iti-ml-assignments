```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
import numpy as np
```

```python

# Create a concentric circle dataset
X, _ = make_circles(n_samples=500, factor=.5, noise=.03, random_state=4)
```

dbscan.png

```python
# Apply DBSCAN to the dataset
dbscan = DBSCAN(eps = 0.2, min_samples=3, metric="euclidean")
labels = dbscan.fit_predict(X)
```

```python
# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=25)
plt.title('DBSCAN Clustering')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.show()
```
